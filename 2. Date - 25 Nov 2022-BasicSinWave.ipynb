Model a wave using mathematics tool

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,201)                                                        #generate 201 point of x in the range of 0 and 20
y = np.sin(x)                                                                 

plt.figure(figsize = (10 , 6))
plt.plot(x,y,'r')                                                                #PLotting x,y and the plot color is red
plt.ylabel('Amplitude')
plt.xlabel('Locaion (x)')
plt.show()
```

The changes over time at specific location x = 2.5 showing as a red Dot.

```python
fig = plt.figure(figsize = (8,8))

times = np.arange(5) # generate a array of 0 to 4.


n = len(times)

for t in times:
  plt.subplot(n, 1, t+1)
  y = np.sin(x+t)
  plt.plot(x,y,'b')
  plt.plot(x[25],y[25],'ro')
  plt.ylim(-1.1,1.1)
  plt.ylabel('y')
  plt.title(f"t = {t}")

plt.xlabel('location (x)')
plt.tight_layout()
plt.show()
```

<b>Characterstics of a wave</b></br><br>

 time period, (T) = 1 / frequency (f)</br><br>

 <ol>A sin wave represented by<br>
 
 <ol>y(t) = A * sin( ω*t + Φ )<br></ol>

 where, A is Amplitude<br>      <ol> and ω = 2 * π / T = 2 * π * f


Generate two sinewave with the time between 0 and 1 seconds. Both waves have frequncy 5hz and 10 hz, Sampled at 100hz but the phase at 0 for both the wave. Also the amplitude of the two waves are 1.Plot two waves and see the differences.<br>

phase, Φ = 0 <br>
Amplitude  = 1<br>

```python
#sampling rate
sr = 100.0
#sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)
# print(t)

#frequency of the signal
freq = 5
y =np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(2,1,1)
plt.plot(t,y,'b')
plt.ylabel('Amplitude')

freq = 10
y =np.sin(2*np.pi*freq*t)


plt.subplot(2,1,2)
plt.plot(t,y,'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```

Generate two sinewave with the time between 0 and 1 seconds. Both waves have frequncy 5hz and sampled at 100hz but the phase at 0 and 10 respective. Also the amplitude of the two wavws are 5 and 10.Plot two waves and see the differences.<br>
frequency , f = 5 Hz<br>
phase, Φ = 0 and 10 <br>
Amplitude  = 5 and 10 <br>

```python
sr = 100.0     #sampling rate
ts = 1.0/ sr    #sampling interval

freq = 5 
phase1 = 0
phase2 = 10
A1 = 5
A2 = 10

t = np.arange(0,1,ts)
y = A1 * np.sin(2*np.pi*freq*t + phase1)
 
plt.figure(figsize = (8,8))
plt.subplot(2,1,1)
plt.plot(t,y,'b')
plt.ylabel('Amplitude')
plt.xlabel('t (in secs)')

y = A2 * np.sin(2*np.pi*freq*t + phase2)
 
plt.figure(figsize = (8,8))
plt.subplot(2,1,2)
plt.plot(t,y,'b')
plt.ylabel('Amplitude')
plt.xlabel('t (in secs)')

plt.show()
```
